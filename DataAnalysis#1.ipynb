## Hypotheses

```python
import scipy.stats as stats
import math
np.random.seed(6)
price = stats.poisson.rvs(loc=18,mu=35,size=1500)
height = stats.poisson.rvs(loc=18,mu=30,size=60)
```

```python
height.mean()
```

```python
_,p_value=stats.ttest_1samp(a=price, popmean=height.mean())
```

```python
price.mean()
```

```python
if p_value<0.05:
    print("we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```
